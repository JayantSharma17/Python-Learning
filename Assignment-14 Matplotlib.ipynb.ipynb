Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
# A scatter plot is a type of data visualization that represents individual data points as dots on a two-dimensional plane.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('My Graph')
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np

# For line 1: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 4, 1)
plt.plot(x,y)
plt.title("For line 1:")

# For line 2: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 4, 2)
plt.plot(x,y)
plt.title("For line 2:")

# For line 3: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 4, 3)
plt.plot(x,y)
plt.title("For line 3:")

# For line 4: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 4, 4)
plt.plot(x,y)
plt.title("For line 4:")

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Ans. A bar plot is a type of data visualization that uses rectangular bars to represent data. 
Each bar in the plot corresponds to a category or group from the dataset, and the length or 
height of the bar represents the value or frequency of the data associated with that category. 

The primary purpose of a bar plot is to:
Comparison
Distribution
Categorical Relationships
Visual Communication

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.title('Company Data Graph')
```

```python
plt.barh(company,profit)
plt.xlabel('profit')
plt.ylabel('company')
plt.title('Company Data Graph')
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
# A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. 
# It displays a summary of the data's central tendency, spread, and potential outliers.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box 1')
```

```python
plt.boxplot(box2)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box 2')
```
