Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans .
1:Line Plot
Use: Line plots are used to show the trend or pattern in a dataset over a continuous interval. 
They are often used for time series data or to visualize how a variable changes over time or another continuous variable.
Function: seaborn.lineplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset
tips = sns.load_dataset('tips')
sns.lineplot(x='total_bill', y='tip', data=tips)
plt.title('Line Plot: Average Tip Amount by Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Average Tip Amount')
plt.show()
```

2:Heatmap
Use: Heatmaps are used to display the correlation between variables in a tabular format. 
It is particularly useful when dealing with a large dataset and looking for patterns and relationships between variables.
Function: seaborn.heatmap()

```python
correlation_matrix = tips.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap: Correlation between Numerical Variables')
plt.show()
```

3:Histogram
Use: Histograms are used to visualize the distribution of a single numerical variable. 
It shows the frequency or count of values within specific intervals (bins).
Function: seaborn.histplot()

```python
sns.histplot(tips['total_bill'])
plt.title('Histogram: Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.show()
```

4:Bar Plot
Use: Bar plots are used to display and compare the distribution of categorical data or 
the relationship between a categorical variable and a numerical variable.
Function: seaborn.barplot()

```python
sns.barplot(x='day', y='tip', data=tips)
plt.title('Bar Plot: Total Tip Amount by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Tip Amount')
plt.show()
```

5:Scatter Plot
Use: Scatter plots are used to visualize the relationship between two continuous variables. 
Each data point is represented as a dot on the graph, and the position of the dot is determined by the values of the two variables.
Function: seaborn.scatterplot()

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot: Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri,hue='event',style='region')
plt.title('Line Plot: Signal vs. Timepoint for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.title('fmri data')
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.boxplot(x='pclass',y='age',data=titanic)
plt.title('Box Plot: Age across Passenger Classes')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Box Plot: Fare across Passenger Classes')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```
